```python
unique_value_counts = data1[object_columns].apply(lambda x: len(x.unique()))
print(len(object_columns))
threshold = 0
print(unique_value_counts)
repetitive_columns = unique_value_counts[unique_value_counts > threshold].index
low_unique_value = unique_value_counts[unique_value_counts <= threshold].index
print("Columns with Repetitive Values:")
print((repetitive_columns))
low_unique_value
```

```python
from sklearn.feature_extraction import FeatureHasher
import pandas as pd

hasher = FeatureHasher(n_features=2, input_type='string')

hashed_dfs1 = []
for column in repetitive_columns:
    data_list = data1[column].apply(lambda x: [str(x)])
    hashed_features = hasher.transform(data_list)
    hashed_df = pd.DataFrame(hashed_features.toarray(), columns=[f'{column}_hash_feature_{i}' for i in range(2)])
    hashed_dfs1.append(hashed_df)

data_encoded1 = pd.concat([data1] + hashed_dfs1, axis=1)

data_encoded1 = data_encoded1.drop(repetitive_columns,axis=1)
data_encoded1.info()
```
